## Malkiel's Bond Pricing Theorems
Malkiel’s theorems formalize the key relationships between bond prices and yields, showing how maturity, coupon level, and the nonlinear nature of discounting combine to govern price sensitivity. Together, they provide the foundation for understanding convexity, term‐structure effects, and the asymmetric impact of yield changes on fixed‐income securities.

* __Source:__ [Malkiel, B. G. (1962). Expectations, Bond Prices, and the Term Structure of Interest Rates. The Quarterly Journal of Economics, 76(2), 197–218. https://doi.org/10.2307/1880816](https://www.jstor.org/stable/1880816)

The five pricing theorems that we'll look at are:
* __Theorem 1:__ Bond prices move inversely to bond yields.
* __Theorem 2:__ For a given change in yield from the nominal yield, changes in bond prices are greater the longer the term to maturity.
* __Theorem 3:__ The percentage price changes described in Theorem 2 increase at a diminishing rate as N increases.
* __Theorem 4:__ Price movements resulting from equal absolute increases and decreases in yield are asymmetric; i.e., decreasing yields raises bond prices more than the same increase in yields lowers prices.
* __Theorem 5:__ The higher the coupon carried by the bond, the smaller the percentage price fluctuation for a given change in yield, except for one-year securities.

## Derivation
Suppose you are a bond trader, and you have a portfolio of bonds that you need to manage. You know that interest rates are going to change, and you want to know how this will affect the value of your portfolio. Suppose the price of the bill, note or bond is denoted as $V_{B}(\bar{r}, \bar{c}, T)$, where $\bar{r}$ is the discount rate, $\bar{c}$ is the coupon rate, and $T$ is the time to maturity:
$$
\boxed{
    dV_{B}
=\underbrace{\frac{\partial V_{B}}{\partial \bar r}}_{V_{B,\bar{r}}}\,d\bar r
+\underbrace{\frac{\partial V_{B}}{\partial \bar c}}_{V_{B,\bar{c}}}\,d\bar c
+\underbrace{\frac{\partial V_{B}}{\partial T}}_{V_{B,T}}\,dT
}
$$
where each partial derivative is defined is evaluated at the current values of $\bar{r}$, $\bar{c}$, and $T$, and $d\bar{r}$, $d\bar{c}$, and $dT$ are the (infinitesimal) changes in the discount rate, coupon rate, and time to maturity, respectively. The function forms for the partial derivatives depends upon the type of Treasury security, particularly, is it is zero-coupon or coupon-bearing, and the type of coupon (fixed or floating).

Let's start with the simplest case, a zero-coupon Treasury bill. 

### Zero-Coupon Treasury Bill
Previously, we showed that the price of a zero-coupon Treasury bill is given by: $V_{B} = \mathcal{D}_{nT,0}^{-1}(\bar{r})\;{V_{P}}$,
where $V_{P}$ is the face (par) value of the bill, and $\mathcal{D}^{-1}_{nT,0}(\bar{r})$ is the (inverse) discrete multiperiod discount factor for compounding events $0\rightarrow{nT}$, which is given by:
$$
\mathcal{D}^{-1}_{nT,0}(\bar{r}) = \left(1+\bar{r}/n\right)^{-nT}
$$
where $n$ is the number of compounding periods per year, and $T$ is the time to maturity in years. The change in the price of the bill with respect to the discount rate $\bar{r}$ and time to maturity $T$ is given by:
$$
\begin{align*}
    dV_{B} & =\frac{\partial V_{B}}{\partial \bar r}\,d\bar{r}+\underbrace{\frac{\partial V_{B}}{\partial \bar c}\,d\bar c}_{=\,0} +\frac{\partial V_{B}}{\partial T}\,dT\\
    &= -\underbrace{V_{P}\left(1+\bar{r}/n\right)^{-nT}\left(\frac{T}{1+\bar{r}/n}\right)}_{V_{B,\bar{r}}}\;d\bar{r} + 0 - \underbrace{V_{P}\left(1+\bar{r}/n\right)^{-nT}\left(n\ln(1+\bar{r}/n)\right)}_{V_{B,T}}\;dT\\
    &= -V_{B,\bar{r}}\;d\bar{r} - V_{B,T}\;dT\quad\blacksquare

\end{align*}
$$
where $V_{B,\bar{r}}$ is the sensitivity of the bond price to changes in the discount rate, and $V_{B,T}$ is the sensitivity of the bond price to changes in time to maturity. Both of these sensitivities are non-negative:
$$
\begin{align*}
V_{B,\bar r}
& =\underbrace{V_{P}\Bigl(1+\tfrac{\bar r}{n}\Bigr)^{-nT}\;\left(\frac{T}{1+\tfrac{\bar r}{n}}\right)}_{\geq\;0}\quad{\text{and}}\quad
V_{B,T}
 = \underbrace{V_{P}\Bigl(1+\tfrac{\bar r}{n}\Bigr)^{-nT}\;\left(n\,\ln\;\Bigl(1+\tfrac{\bar r}{n}\Bigr)\right)}_{\geq\;0},
\end{align*}
$$
From these expressions, we can see the direction of the price changes with respect to changes in the discount rate and time to maturity.  Let's take a closer look at how the price of the bill changes with respect to these two variables.
* __Inversely proportional to the discount rate:__ As the discount rate increases, the price of the bill decreases, and vice versa. This is because a higher discount rate means that future cash flows are worth less in present value terms. So $V_{B}\downarrow$ as $\bar{r}\uparrow$, assuming $dT\approx 0$.

    > __Example:__ You buy a 52-week T-bill at auction. If the discount rate increases from 2% to 3%, the price in the secondary market will be **lower** than your purchase price, because the future cash flow (the face value) is discounted more heavily.

* __Inversely proportional to time to maturity:__ As the time to maturity increases, the price of the bill decreases for the same discount rate, and vice versa.
Thus, $V_{B}\downarrow$ as $T\uparrow$, assuming $d\bar{r}\approx 0$.  However, **in real time** as you hold the bill, calendar time $t$ advances and the remaining time $T$ **decreases** ($dT<0$):

   > **Example:** You buy a 52-week T-bill at auction. Six months later ($T$ falls from 1 yr to 0.5 yr) and with $\bar r$ unchanged, the price in the secondary market will be **higher** than your purchase price—exactly because there are fewer remaining periods to discount.

Now, let's consider a coupon-bearing Treasury note or bond.

### Coupon-Bearing Treasury Note or Bond
Previously, we showed that the price of a coupon-bearing Treasury note (bond) is given by:
$$V_{B} = \mathcal{D}^{-1}_{nT,0}(\bar{r})\cdot{V_{P}}+C\cdot\sum_{j=1}^{n{T}}\mathcal{D}^{-1}_{j,0}(\bar{r})$$
where $C\equiv(\bar{c}/n)\cdot{V_{P}}$ is the value of a coupon payment, $\bar{c}$ is the coupon rate, and $\mathcal{D}^{-1}_{j,0}(\bar{r}) = \left(1+\bar{r}/n\right)^{-j}$ is the discrete multiperiod discount factor for periods $0\rightarrow{j}$. The change in the price of the note (bond) with respect to the discount rate $\bar{r}$, coupon rate $\bar{c}$, and time to maturity $T$ is given by:
$$
\begin{align*}
    dV_{B} & =\frac{\partial V_{B}}{\partial \bar r}\,d\bar{r}+\frac{\partial V_{B}}{\partial \bar c}\,d\bar c +\frac{\partial V_{B}}{\partial T}\,dT\\
    &= -\underbrace{V_{P}(1+\bar{r}/n)^{-nT}\left(\frac{nT}{1+\bar{r}/n}\right)\left(\frac{1}{n}\right) + C\sum_{j=1}^{nT}(1+\bar{r}/n)^{-j}\left(\frac{j}{1+\bar{r}/n}\right)\left(\frac{1}{n}\right)}_{V_{B,\bar{r}}}\;d\bar{r}\\
    &\quad + \underbrace{\frac{V_{P}}{n}\sum_{j=1}^{nT}(1+\bar{r}/n)^{-j}}_{V_{B,\bar{c}}}\;d\bar{c}\\
    &\quad + \underbrace{V_{P}(1+\bar{r}/n)^{-nT}(-n\ln(1+\bar{r}/n)) + C(1+\bar{r}/n)^{-nT}(-n\ln(1+\bar{r}/n))}_{V_{B,T}}\;dT\\
    &= -V_{B,\bar{r}}\;d\bar{r} + V_{B,\bar{c}}\;d\bar{c} + V_{B,T}\;dT\quad\blacksquare
\end{align*}
$$
where $V_{B,\bar{r}}$ is the sensitivity of the bond price to changes in the discount rate, $V_{B,\bar{c}}$ is the sensitivity to changes in the coupon rate, and $V_{B,T}$ is the sensitivity to changes in time to maturity. All of these sensitivities are non-negative:
$$
\begin{align*}
V_{B,\bar r} &= \underbrace{V_{P}(1+\tfrac{\bar r}{n})^{-nT}\left(\frac{T}{1+\tfrac{\bar r}{n}}\right) + C\sum_{j=1}^{nT}(1+\tfrac{\bar r}{n})^{-j}\left(\frac{j}{n(1+\tfrac{\bar r}{n})}\right)}_{\geq\;0}\\
V_{B,\bar c} &= \underbrace{\frac{V_{P}}{n}\sum_{j=1}^{nT}(1+\tfrac{\bar r}{n})^{-j}}_{\geq\;0}\\
V_{B,T} &= \underbrace{-V_{P}(1+\tfrac{\bar r}{n})^{-nT}n\ln(1+\tfrac{\bar r}{n}) - C(1+\tfrac{\bar r}{n})^{-nT}n\ln(1+\tfrac{\bar r}{n})}_{\leq\;0}
\end{align*}
$$
From these expressions, we can see the direction of the price changes with respect to changes in the discount rate, coupon rate, and time to maturity. 

Let's examine each sensitivity:

* __Inversely proportional to the discount rate:__ As the discount rate increases, the price of the note (bond) decreases, and vice versa. This effect is stronger for longer-maturity bonds because more future cash flows are discounted. So $V_{B}\downarrow$ as $\bar{r}\uparrow$, assuming $d\bar{c}\approx 0$ and $dT\approx 0$.

    > __Example:__ You buy a 10-year Treasury note at auction with a 3% coupon rate. If the discount rate increases from 3% to 4%, the price in the secondary market will be **lower** than your purchase price, because both the coupon payments and principal repayment are discounted more heavily.

* __Directly proportional to the coupon rate:__ As the coupon rate increases, the price of the note (bond) increases, and vice versa. Higher coupon payments increase the present value of the bond. So $V_{B}\uparrow$ as $\bar{c}\uparrow$, assuming $d\bar{r}\approx 0$ and $dT\approx 0$.

    > __Example:__ Compare two 10-year Treasury notes with identical discount rates but different coupon rates (2% vs. 4%). The bond with the higher coupon rate will have a **higher** price because it provides larger periodic cash flows.

* __Inversely proportional to time to maturity:__ As the time to maturity increases, the price of the note (bond) decreases for the same discount rate and coupon rate, and vice versa. However, the mathematical relationship is more complex than for zero-coupon bonds. The sensitivity $V_{B,T}$ is actually negative, indicating that as $T$ increases, the bond price decreases. In **real time** as you hold the bond, calendar time $t$ advances and the remaining time $T$ **decreases** ($dT<0$), so the negative sensitivity means the price increases over time (all else equal):

   > **Example:** You buy a 10-year Treasury note at auction. Five years later ($T$ falls from 10 yr to 5 yr) and with $\bar r$ and $\bar{c}$ unchanged, the price in the secondary market will be **higher** than your purchase price—the negative time sensitivity combined with the negative change in time ($dT<0$) results in a positive price change.
